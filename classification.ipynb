# Classifiers
In order to create an algorithm to classify, I first wanted to create a [Classifier Problem Space Generator](https://github.com/bobaba/classifiers/blob/master/cps_generator.py).

This Generator accepts the following inputs:  
1. **lengthOfSides** establishes the lengths of x and y in the plot, also sets bounds for randomized points  
2. **numPoints** sets the number of randomized points to plot  
3. **numClasses** each randomly created point will also receive a randomized class (1-*numClasses*)  
4. **numUnseen** creates *n* number of points without classes  

Instantiated, this will provide the following methods:  
*object*.**data** contains \[*x,y,c*\] where *x*=x-axis, *y*=y-axis, and *c*=class assigned  
*object*.**unseen** contains \[*x,y*\] where *x*=x-axis, and *y*=y-axis, with no class assigned  
*object*.**showScatterPlot()** will display the plot created with Matplotlib  

After I got the generator working, I was then able to code the [K-Nearest Neighbor Classifier Algorithm]() to attempt solving these randomized plots. It runs quickly and efficiently, but is not always accurate.

```python
from cps_generator import CPSGenerator
from k_nearest_neighbor_classifier import KNearestNeighbor
# Set up Problem Space
lengthOfSides = 20
numPoints = 150
numClasses = 3
numUnseen = 3
a = CPSGenerator(lengthOfSides,numPoints,numClasses,numUnseen)

# Solve with k-Nearest Neighbor
knn = KNearestNeighbor(a.data,a.unseen, k=5)
knn.prettyPrintResults()
a.showScatterPlot()
```
